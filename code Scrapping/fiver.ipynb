```python
import csv
from selenium import webdriver
```

```python


driver = webdriver.Chrome()
driver.get("https://www.fiverr.com/categories/programming-tech/website-development?context=recommendation&context_referrer=tailored_homepage_perseus&source=gigs_cold_start_by_sc")
```

```python
# Get all freelancers
freelancers = driver.find_elements("xpath", "//div[contains(@class, 'gig-card-layout')]")

# list to store freelancers data
freelancers_data = []

for freelancer in freelancers:
    data = {}

    # Descrption
    description = freelancer.find_elements("xpath", ".//a[@class='do6Dc_f FSxhdEL']/p")
    data["Description"] = description[0].text if description else "Not found"

    # Name
    name = freelancer.find_elements("xpath", ".//span[@class='_1sz16f5k _1sz16f5l z58z8724z z58z8725a z58z872 z58z871hc z58z87a z58z871i6']")
    data["Name"] = name[0].text if name else "Not found"

    # rate
    rate = freelancer.find_elements("xpath", ".//p[@class='z58z872']")
    data["Rate"] = rate[0].text if rate else "Not found"

    # Number of stars
    stars = freelancer.find_elements("xpath", ".//strong[@class='rating-score ZkCIs6P']")
    data["Stars"] = stars[0].text if stars else "Not found"

    # Number of reviewss
    reviews = freelancer.find_elements("xpath", ".//span[@class='rating-count-number']")
    data["Reviews"] = reviews[0].text if reviews else "Not found"

    # price
    price = freelancer.find_elements("xpath", ".//span[@class='text-bold co-grey-1200']")
    data["Price"] = price[0].text if price else "Not found"

    # Does he offer video consultations?
    offer_video = freelancer.find_elements("xpath", ".//p[@class='_1sz16f5k z58z871f4 z58z871cm z58z877 z58z872']")
    data["Offers Video Consultation"] = offer_video[0].text if offer_video else "Not found"

    freelancers_data.append(data)

for i, freelancer in enumerate(freelancers_data, 1):
    print(f"\n *Freelancer {i}:*")
    for key, value in freelancer.items():
        print(f"   {key}: {value}")
```

```python
# Save data to a CSV file
csv_file = "fiver_data.csv"
csv_columns = ["Description", "Name", "Rate", "Stars", "Reviews", "Price", "Offers Video Consultation"]

with open(csv_file, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=csv_columns)
    writer.writeheader()
    writer.writerows(freelancers_data)

print(f"Data saved to {csv_file} successfully!")
```
